```c
#include<stdio.h>
int main(){

int i,n,F;
float S;
printf("Please, give n\n");
scanf("%d",&n);
S=0;
F=1;
for(i=1;i<=n;i++){
S+=(float)i/(i+1);
F*=i;
}
printf("The sum of 1/2+2/3+...+%d/%d=%.4f\n",n,n+1,S);
printf("Factorial(%d)=1x2x....x%d=%d\n",n,n,F);
return 0;}
```
#### Recursive function
```c
#include<stdio.h>
float S(int n);
int F(int n);

int main(){
int i,n;
do{printf("Please, give n\n");
scanf("%d",&n);
}while(n<=0);
printf("The sum of 1/2+2/3+...+%d/%d=%.4f\n",n,n+1,S(n));
printf("Factorial(%d)=1x2x....x%d=%d\n",n,n,F(n));
return 0;}

float S(int n){
if (n==1) return 1/2.;
else return S(n-1)+(float)n/(n+1);
}

int F(int n){
if (n==0) return 1;
else return n*F(n-1);
}
```

### Exercise 6
```c
#include<stdio.h>
int quotient(int,int);
int reminder(int,int);
int main(){
int a,b;
printf("Please give a and b");
scanf("%d %d",&a,&b);
printf("The quotient is %d and the reminder is %d",quotient(a,b),reminder(a,b));
return 0;}

int quotient(int a,int b){
if(a<b){return 0;}
else{return 1+quotient(a-b,b);}
}

int reminder(int a,int b){
if(a<b){return a;}
else{return reminder(a-b,b);}
}
```
