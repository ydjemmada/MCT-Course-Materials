#### Exercise 4
Suppose there is a typed file called "exam.txt" that contains records related to candidates in a competition. Each record is composed of: ID, NAME, FIRST NAME, AGE, DECISION (a type containing the following identifiers: admitted, rejected, deferred), separated by semicolons (;).

Required work:

- Define the `add()` function that allows filling in the data related to the candidates in the `exam.txt` file.
- Define the `success()` function that allows creating the `success.txt` file containing data related to admitted candidates.
- To prioritize admitted candidates who are under 30 years old, create the `waiting()` function that produces a new file called `waiting.txt` containing data related to admitted candidates who are over 30 years old. A line in the `waiting.txt` file includes the ID, NAME, and FIRST NAME of a candidate separated by semicolons (;).
- Define the `statistics(dec)` function that allows returning the percentage of candidates for the decision `dec` (admitted, rejected, and deferred). For example: $$\text{the percentage of admitted candidates} = \text{(Number of admitted candidates / Total number of candidates)}\times 100.$$

- Define the `delete()` function that will delete candidates over 30 years old from the `success.txt` file.

```python
#Exercise 04
def add():
    decision={"a":"admitted", "r":"rejected", "d":"deferred"}
    try:
        f = open("exam.txt",'r')
    except FileNotFoundError:
        f = open("exam.txt", 'a+')
        print("File Created")
    existing_ids=[line.split(';')[0] for line in f]
    while True:
        ID = input("Enter candidate ID: ")
        if ID in existing_ids:
            print("\nID already exist!\n")
        else:
            break
    name = input("Enter candidate name: ")
    first_name = input("Enter candidate first name: ")
    while True:
        age = input("Enter candidate age: ")
        if int(age) not in range(1,100):
            print("\nWrong age\n")
        else:
            break
    while True:
        dec = input("Enter candidate decision (a:admitted, r:rejected, d:deferred): ")
        if dec not in decision.keys():
            print("\nWrong decision\n")
        else:
            break
    f.close()
    with open("exam.txt",'a') as f:
        f.write(ID+";"+name+";"+first_name+";"+age+";"+decision[dec]+";\n")
def success():
    with open("exam.txt",'r') as f:
        with open("success.txt",'w+') as f1:
            for line in f:
                if line.split(';')[4]=="admitted":
                    f1.write(line)
def waiting():
    with open("success.txt",'r') as f:
        with open("waiting.txt",'w+') as f1:
            for line in f:
                l=line.split(';')
                if int(l[3])>=30:
                    f1.write(l[0]+';'+l[1]+';'+l[2]+';\n')
def statistics(dec):
    with open("exam.txt",'r') as f:
        total=0
        s=0
        for line in f:
            total+=1;
            l=line.split(';')
            
            if dec==l[4] and dec in ["admitted","rejected","deferred"]:
                s+=1
            if dec not in ["admitted","rejected","deferred"]:
                return -1
    return (s/total*100)
def delete():
    with open("success.txt",'r') as f:
        lines=f.readlines()
        to_write_lines=[]
        for line in lines:
            if int(line.split(';')[3])<30:
                to_write_lines.insert(0,line)
    with open("success.txt",'w') as f:
        f.writelines(to_write_lines)
add()
success()
waiting()
statistics("rejected")
delete()
```

```python
import string
# 1. Read in the contents of the file and display them on the screen.
with open("sample.txt", "r") as f:
    contents = f.read()
    print("File Contents:\n", contents)

# 2. Count the number of occurrences of each word in the file and display the results on the screen.
words = contents.lower().split()
word_count = {}
for word in set(words):
    word = word.lower()
    word_count[word]=words.count(word)
print("\nWord count:\n", word_count)
# 3. Find the most common word in the file and display it on the screen.
most_common_word={word:word_count[word] for word in word_count if word_count[word]==max(word_count.values())}
print("\nMost common word:\n", most_common_word)
# 4. Find the longest word in the file and display it on the screen.
print(max(contents.split(),key=len))
print(max([word.strip(string.punctuation) for word in words], key=len))
without_vowels=[char for char in contents if char not in "aeuioAEUIO"]
print("\nText without voyels is:\n", "".join(without_vowels))
```
