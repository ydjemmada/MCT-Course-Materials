```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define parameters
n = 200
theta = np.linspace(0, 2*np.pi, n)
w = 3
phi = 0.5 * theta

# Define parametric equations for the Möbius strip
x = (w + np.cos(phi)) * np.cos(theta)
y = (w + np.cos(phi)) * np.sin(theta)
z = np.sin(phi)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the Möbius strip
ax.plot(x, y, z, color='r')

# Set the limits of the plot
ax.set_xlim([-4, 4])
ax.set_ylim([-4, 4])
ax.set_zlim([-2, 2])

# Set the labels of the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.show()
```
