## Definitions
### Permutations
```latex

\begin{block}{}
A \textbf{Permutation} $\sigma$ of the set $[n]=\{1,2,\dots,n\}$ is a rearrangement (linear ordering) of its elements.
\end{block}
```
### Partitions

```latex
\begin{block}{}
A \textbf{partition} $\pi$ of the set $[n]=\{1,2,\dots,n\}$ into $k$ parts is a distribution of its elements into $k$ nonempty disjoint subsets $b_1,b_2,\dots,b_k$ called \textit{blocks}. We denote
\begin{equation*}
\pi([n])=b_1|b_2|\dots|b_k,
\end{equation*}
with $\bigcup_{i=1}^{k}b_i=[n]$ and $b_i \neq \emptyset$, for $\ i \in \{1,\dots,k\}$.
\end{block}
```

### Stirling numbers of the second kind

\begin{block}{}
  \textbf{Stirling numbers of the second kind}, denoted $ {n \brace k}$, count the number of all partitions of $[n]$ having exactly $k$ blocks.
\end{block}
