#### Recursive function
```c
#include<stdio.h>
float S(int n);
int F(int n);

int main(){
int i,n;
do{printf("Please, give n\n");
scanf("%d",&n);
}while(n<=0);
printf("The sum of 1/2+2/3+...+%d/%d=%.4f\n",n,n+1,S(n));
printf("Factorial(%d)=1x2x....x%d=%d\n",n,n,F(n));
return 0;}

float S(int n){
if (n==1) return 1/2.;
else return S(n-1)+(float)n/(n+1);
}

int F(int n){
if (n==0) return 1;
else return n*F(n-1);
}
```

````python
###
#
# Read the list of medical images 
# Create 3-prngs
# Encrypt using XOR
# Decrypt using the same function
# Safe the encrypted Image
###

import sympy as sp
def lehmer_rng(seed,a,b,m):
    return np.mod(a*seed+b,m)
def encrypt_decrypt(img,seed,a,b,m):
    for l in range(3):
        for i in np.arange(img.shape[0]):
            for j in np.arange(img.shape[1]):
                a=sp.nextprime(a)
                b=sp.nextprime(b)
                seed=lehmer_rng(seed,a,b,m)
                img[i,j,l]=np.bitwise_xor(img[i,j,l],seed)    
    return img
img=plt.imread("med_1.jpg")
plt.imshow(img)
plt.show()
img=encrypt(img,473,27,300,274**2)
plt.imshow(img)
plt.show()
plt.imshow(encrypt(img,473,27,300,274**2))
````

````python
#Solution of the Covid project
import os
import numpy as np
import matplotlib.pyplot as plt
def annotate(img,state):
    hight=img.shape[0]
    width=img.shape[1]
    if state==0:
        img[0:5,:,:]=[0,128,0]
        img[-5:,:,:]=[255, 165, 0]
        img[:,0:5,:]=[255, 165, 0]
        img[:,-5:,:]=[255, 165, 0]
    elif state==1:
        img[0:5,:width//2,:]=[128,0,0]
        img[-5:,:width//2,:]=[128,0,0]
        img[:,0:5,:]=[128,0,0]
        img[:,width//2:width//2+5,:]=[128,0,0]
    elif state==2:
        img[0:5,width//2:,:]=[0,0,128]
        img[-5:,width//2:,:]=[0,0,128]
        img[:,width//2:width//2+5,:]=[0,0,128]
        img[:,-5:,:]=[0,0,128]
    return img
location = os.getcwd()+"/test"
num=1

for filename in os.listdir(location):
    img=plt.imread(os.path.join(location,filename))
    test=np.count_nonzero(img>=128)/img.size*100
    num+=1
    if(50<test and test<=65):
        plt.imsave('results/result'+str(num)+'.png',annotate(img,0))
    elif(test>65):
        plt.imsave('results/result'+str(num)+'.png',annotate(img,1))
    else:
        plt.imsave('results/result'+str(num)+'.png',annotate(img,2))
````

````python
#Hungarian method
import matplotlib.pyplot as plt
import numpy as np
import pandas
def Mark(A):
    steps=0
    n=A.shape[0]
    M=np.zeros((n,n))
    # Step 1: boxed zeros and barred zeros
    while(np.count_nonzero(M==1)!=n):
        M=np.zeros((n,n))
        steps+=1
        num_zeros = (M == -1).sum(axis=1)
        i_set = np.argsort(num_zeros)
        while (True):
            zero_boxed=False
            for i in i_set:
                for j in np.arange(n):
                    if(A[i,j]==0. and M[i,j]==0.):
                        M[i,j]=1
                        zero_boxed=True
                        for k in np.arange(j+1,n):
                            if(A[i,k]==0 and M[i,k]!=1):
                                M[i,k]=-1
                        for k in np.arange(i+1,n):
                            if(A[k,j]==0 and M[k,j]!=1):
                                M[k,j]=-1
            if np.sum(M==1)==n:
                return A,M
            if(not zero_boxed):
                break
        new_arr = np.where(M == 1, '_', np.where(M == -1, '*', A))
        print("\nIteration ",steps,"+-------+\n")
        print("Step1: boxed zeros and barred zeros\n")
        print(pandas.DataFrame(new_arr))
        # Step 2: Mark and color row/columns
        print("Step 2: marked_colored\n")
        marked_rows=np.zeros(n,dtype=bool)
        marked_cols=np.zeros(n,dtype=bool)
        raw_mark_test=False
        for i in np.arange(n):
            if 1. not in M[i,:]:
                marked_rows[i]=True
                raw_mark_test=True
        while(raw_mark_test==True):
            raw_mark_test=False
            for i in np.where(marked_rows==True)[0]:
                for j in np.arange(n):
                    if (M[i,j]==-1 and marked_cols[j]!=True):
                        marked_cols[j]=True
            for i in np.arange(n):
                for j in np.where(marked_cols==True)[0]:
                    if (M[i,j]==1 and marked_rows[i]!=True):#Not already marked
                        marked_rows[i]=True
                        raw_mark_test=True
        colored_rows=np.where(marked_rows==False)[0]
        colored_cols=np.where(marked_cols==True)[0]
        not_colored_rows=np.where(marked_rows==True)[0]
        not_colored_cols=np.where(marked_cols==False)[0]
        print("\ncolored_rows {} colored_cols {}".format(colored_rows,colored_cols))
        
        # Step 3: Update the matrix +/-
        print("\nStep 3:update coast matrix\n")
        min_value=np.min(A[not_colored_rows][:,not_colored_cols])
        for i in not_colored_rows:
            for j in not_colored_cols:
                    A[i,j]-=min_value
        for i in colored_rows:
            for j in colored_cols:
                    A[i,j]+=min_value
        print(pandas.DataFrame(A))
    return (A,M)                

print("\nCoast Matrix=")
coasts=np.array(
[[17,15,9,4,12],
[16,16,10,5,10],
[12,15,14,11,5],
[4,8,14,17,13],
[13,9,8,12,17]])
#Step 0
print(pandas.DataFrame(coasts))

print("Step 0: Matrix_reduction\n")
A=coasts.copy()
for i in np.arange(A.shape[0]):
    A[i,:]=A[i,:]-np.min(A,axis=1)[i]
for i in np.arange(A.shape[0]):
    A[:,i]=A[:,i]-np.min(A,axis=0)[i]
print(pandas.DataFrame(coasts))
#Calling the Steps 1-3
A,M=Mark(A)

print("\nSolution ******\n")
print("\nCoast Matrix=\n",coasts)


for i,j in np.stack(np.where(M==1),axis=1):
    print(("employee:",i,"task:",j))
print("the coasts are:",coasts[M==1])

print("the total coast is:",sum(coasts[M==1]))
````
